```python
import pandas as pd
from pathlib import Path
```

```python
data_folder = Path("../data/raw")
```

```python
files = [
    "ucl_2018_19.csv",
    "ucl_2019_20.csv",
    "ucl_2020_21.csv",
    "ucl_2021_22.csv",
    "ucl_2022_23.csv",
    "ucl_2023_24.csv"
]
```

```python
season_dfs = []
```

```python
for file in files:
    file_path = data_folder / file
    df = pd.read_csv(file_path, encoding='latin1')
    df = df[['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']]
    df = df.dropna()
    df = df.drop_duplicates()
    df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
    season_dfs.append(df)
```

```python
all_matches = pd.concat(season_dfs, ignore_index=True)
```

```python
all_matches = all_matches.sort_values(by='Date').reset_index(drop=True)
```

```python
df = pd.read_csv("../data/combined_ucl_matches.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.sort_values(by='Date').reset_index(drop=True)
```

```python
home_df = df[['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG']].copy()
home_df.columns = ['Date', 'Team', 'Opponent', 'GoalsFor', 'GoalsAgainst']         
```

```python
away_df = df[['Date', 'AwayTeam', 'HomeTeam', 'FTAG', 'FTHG']].copy()
away_df.columns = ['Date', 'Team', 'Opponent', 'GoalsFor', 'GoalsAgainst']
```

```python
team_df = pd.concat([home_df, away_df], ignore_index=True)
```

```python
team_df = team_df.sort_values(by='Date').reset_index(drop=True)
```

```python
team_df.shape
```

```python
team_df['Result'] = 0
team_df.loc[team_df['GoalsFor'] > team_df['GoalsAgainst'], 'Result'] = 1
team_df.loc[team_df['GoalsFor'] < team_df['GoalsAgainst'], 'Result'] = -1
```

```python
team_df['Win'] = (team_df['Result'] == 1).astype(int)
```

```python
team_df['GF_avg_last5'] = (
    team_df.groupby('Team')['GoalsFor']
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)
```

```python
team_df['GA_avg_last5'] = (
    team_df.groupby('Team')['GoalsAgainst']
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)
```

```python
team_df['Wins_last5'] = (
    team_df.groupby('Team')['Win']
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).sum())
)
```

```python
team_df[['Date', 'Team', 'GoalsFor', 'GoalsAgainst', 'Win', 'GF_avg_last5', 'Wins_last5']].head(15)
```

```python
home_features = team_df[['Date', 'Team', 'GF_avg_last5', 'GA_avg_last5', 'Wins_last5']].copy()
home_features.columns = ['Date', 'HomeTeam', 'Home_GF_avg_last5', 'Home_GA_avg_last5', 'Home_Wins_last5']
```

```python
away_features = team_df[['Date', 'Team', 'GF_avg_last5', 'GA_avg_last5', 'Wins_last5']].copy()
away_features.columns = ['Date', 'AwayTeam', 'Away_GF_avg_last5', 'Away_GA_avg_last5', 'Away_Wins_last5']
```

```python
match_features = df.merge(home_features, on=['Date', 'HomeTeam'], how='left')
```

```python
match_features = match_features.merge(away_features, on=['Date', 'AwayTeam'], how='left')
```

```python
match_features.head()
```

```python
match_features.columns
```

```python
match_features[['Home_GF_avg_last5', 'Home_GA_avg_last5', 'Home_Wins_last5', 'Away_GF_avg_last5', 'Away_GA_avg_last5', 'Away_Wins_last5']].isnull().sum()
```

```python
match_features.isnull().sum()
```

```python
feature_cols = [
    'Home_GF_avg_last5', 'Home_GA_avg_last5', 'Home_Wins_last5', 'Away_GF_avg_last5', 'Away_GA_avg_last5', 'Away_Wins_last5'
]
match_features[feature_cols] = match_features[feature_cols].fillna(0)
```

```python
model_df = match_features [['Date', 'HomeTeam', 'AwayTeam', 'Home_GF_avg_last5', 
                            'Home_GA_avg_last5', 'Home_Wins_last5', 'Away_GF_avg_last5', 'Away_GA_avg_last5', 'Away_Wins_last5', 'FTR']].copy()
```

```python
model_df.head(
)
```

```python
model_df.shape
```

```python
model_df['Target'] = model_df['FTR'].map({
    'H': 0,
    'D': 1,
    'A': 2
})
```

```python
split_index = int(len(model_df) * 0.8)

train_df = model_df.iloc[:split_index]
test_df = model_df.iloc[split_index:]
```

```python
match_features.columns
train_df.columns
model_df.columns
```

```python
features = [    
    'Home_GF_avg_last5', 'Home_GA_avg_last5', 'Home_Wins_last5', 'Away_GF_avg_last5', 'Away_GA_avg_last5', 'Away_Wins_last5'
]
X_train = train_df[features]
y_train = train_df['Target']

X_test = test_df[features]
y_test = test_df['Target'] 
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
y_test.value_counts(normalize=True)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
